### Question 1

How many people whose credit purpose is education and property is real estate?

```python
len(df[(df.purpose == "education") & (df.property == "real estate")])
```

### Question 2

What's the credit history of the guy who has the maxium amount among the whole dataset? Your code must return <b>only one string</b>.

```python
df[df.amount == df.amount.max()]['credit_history'].values[0]
```

### Question 3

What is the proportion of people (0 to 1) who are younger than 30 and have lower than 24 months loan duration?

```python
len(df[(df.age < 30) & (df.months_loan_duration < 24)]) / (len(df) + 0.0)
```

### Question 4

For each employment_length, report how many of those records have over 30 months loan duration. You must report one entry for each employment_length.

```python
df2 = df.copy()
df2['Heavy_Loan'] = (df2.months_loan_duration >= 30) + 0.0
df2.groupby('employment_length')['Heavy_Loan'].sum()
```

### Question 5

Report the number of differrent purposes in descending order.

```python
df.groupby(['purpose'])['purpose'].count().sort_values(ascending=False)
```

### Question 6

How do you screen out rows that have the same age as the previous row?

```python
df.loc[df['age'].shift() == df['age']]
```

### Question 7

For each combination of <i>employment_length</i> and <i>job</i>, what's the average, minimum, and maximum number of amount?

```python
df.groupby(['employment_length','job'])['amount'].agg(['mean','min','max'])
```

### Question 8

Let us define "average amount of an age" as the average amount of that age. Find the age group of which the average amount is 20% higher than the average amount of all records. Report both the age and its average amount.

```python
df3 = df.groupby('age')['amount'].mean()
df3[df3 > df.amount.mean() * 1.2]
```

### Question 10

<p>We assume that there may be some correlation between credit history and job. So let us suppose that the one that could be <I>fully repaid</I> should be skilled employee.</p>
<p>For each <I>purpose</I>, find the proportion of people who 'fully repaid' and also are 'skilled employee' <b>among all 'fully repaid'</b></u>.</p>
<p>Your code must return one entry for each purpose with fully repaid. To identify fully repaid, you must check whether credit history <b>contains</b> the word 'fully repaid'; to identify whether a job is skilled employee, you must check whether job contains the word 'skilled employee'. <i>Hint</i>: make a copy of the original data frame before solving this problem.</p>

```python
helper = df[df.credit_history.str.contains('fully repaid')].copy()
helper['skilled_employee'] = helper.job.apply(lambda s : 1.0 if 'skilled employee' in s else 0.0)
helper.groupby('purpose')['skilled_employee'].mean().sort_values(ascending=False)
```
